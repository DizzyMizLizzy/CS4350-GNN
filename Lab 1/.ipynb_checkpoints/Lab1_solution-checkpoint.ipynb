## Noisy measurements

Let $e$ be the average energy of the graph signals in the dataset.
$$
e = \frac{1}{m} \sum_{i=1}^m \|\textbf{x}_i\|_2^2,
$$
where $m$ is the number of data points.
Generate $m$ noisy graph signal $\textbf{y} = \textbf{x} + \textbf{n}$ where $\textbf{n}$ is a zero-mean Gaussian noise $\textbf{n} \sim \mathcal{N}(0,Ïƒ^2\textbf{I})$.
Consider $\sigma^2 = e/2$.

```python
## loading the data
timeseries_data = pd.read_csv('lab1_molene/filtered_data_Molene.csv', index_col = 'date', parse_dates = True)
timeseries_data = timeseries_data - timeseries_data.mean() # => removing mean from data
```

```python
x = timeseries_data.to_numpy().transpose()  # => transpose the data so we have each column as a graph signal

e = np.sum(np.square(x))/x.shape[1] # computing average energy

sigma = e/x.shape[0]/2
n = np.random.multivariate_normal(np.zeros(x.shape[0]),sigma*np.eye(x.shape[0]),size=x.shape[1]).transpose() # => generating noise

y = x + n   # adding to noise to synthesize noisy observations
```

### Task 2

Use the Tikhonov regularizer for a choice of $\alpha > 0$ to recover signal $\textbf{x}$ from the measurements $\textbf{y}$.
Provide the signal recovery metric root normalized mean square error (rNMSE)
$$
\text{rNMSE} = \left(\frac{\sum_{i=1}^m \|\hat{\textbf{x}}_i-\textbf{x}_i\|_2^2}{\sum_{i=1}^m\|\textbf{y}_i\|_2^2}
\right)^{0.5}
$$

```python
alpha = 10

L = G.L.todense()   # making L dense instead of sparse for further calculations in numpy

x_hat = np.matmul(np.linalg.inv(alpha*L+np.eye(L.shape[0])),y) # the closed form solution for Tikhonov regularizer problem

rnmse = np.sqrt(np.square(x_hat-x).mean())/y.std()  # rnmse definition

print(f"The rNMSE is: {np.round(rnmse,4)}")
```

### Question 2

Illustrate rNMSE as a function of $\alpha \in [0,10]$. What do you conclude about the role of the regularizer?




```python
alpha = np.linspace(0,10,100)

rnmse = []
for i in alpha:
  x_hat = np.matmul(np.linalg.inv(i*L+np.eye(L.shape[0])),y)
  rnmse_tmp = np.sqrt(np.square(x_hat-x).mean())/y.std()
  rnmse.append(rnmse_tmp)


plt.plot(alpha,rnmse)
plt.grid()

print(f"The optimum value for the regularizer coefficient is alpha = {np.round(alpha[rnmse.index(min(rnmse))],2)} with rNMSE = {np.round(min(rnmse),4)} ")
```

Answer: The regularizer improves the performance as it enforces a prior knowledge on the data. There is also an optimum value for the regularizer coefficient.

### Task 3

Implement another regularizer of your choice and compare the performance with the Tikhonov regularizer for $\sigma^2 = e/2$.

**Trend filter**

This regularizer tries to compute the sparsity of signal's difference over the graph.
The trend filtering problem is
$$
\hat{\textbf{x}} = \text{arg}\min_{\textbf{x}} \|\textbf{y} - \textbf{x}\|_2^2 + \beta \|\textbf{B}^T\textbf{x}\|_1,
$$
where $\alpha > 0$ is the regularizer coefficient and $\textbf{B}$ is the incidence matrix of the graph performing difference operation.


```python
## Computing differnce operator = B^T
G.compute_differential_operator()
D = G.D.todense()
```

```python
beta = 2
m = 48

x_tf = np.zeros((x.shape[0],m))
for i in range(m):
    x_var = cp.Variable(G.N)

    obj = cp.Minimize(cp.sum_squares(y[:,i] -x_var) + beta * cp.pnorm(D @ x_var, 1))
    prob = cp.Problem(obj)

    prob.solve(solver=cp.CVXOPT, verbose = False)

    if prob.status != cp.OPTIMAL:
        raise Exception("Solver did not converge!")

    x_tf[:,i] = x_var.value

rnmse_tf = np.sqrt(np.square(x_tf-x[:,:m]).mean())/y[:,:m].std()
print(f"For beta = {beta}, the rNMSE is {np.round(rnmse_tf,4)}.")
```

### Question 4

Why and when does one regularizer performs better than the other?

**Answer**:
Tikhonov performs better since the data on the graph is smooth rather than piecewise smooth. A regularizer works better when that data better satisfies its assumptions.

Taking the best regularization weights from the above exercise, show the rNMSE of the recovered signal for both regularizers and $M = \{3,9,15,21,27\}$.

```python
M = [3,9,15,21,27]
indices = []
for m in M:
    indices_tmp = np.sort(np.random.choice(G.N,m,replace=False))
    indices.append(indices_tmp) 
```

```python
alpha = 1.11
rnmse_rc = np.zeros(len(indices))
for i in range(len(indices)):
    x_mv = np.zeros(x.shape)
    x_mv[indices[i],:] = np.copy(x[indices[i],:])
    x_rc = np.matmul(np.linalg.inv(alpha*L+np.eye(L.shape[0])),x_mv)
    rnmse_rc[i] = np.sqrt(np.square(x_rc[indices[i],:]-x[indices[i],:]).mean())/x_mv[indices[i],:].std()
    print(f"The rNMSE for {M[i]} observation is: {np.round(rnmse_rc[i],4)}")
```

```python
beta = 2
m = 48
rnmse_rc_tf = np.zeros(len(indices))
for j in range(len(indices)):
    x_mv = np.zeros((x.shape[0],m))
    x_mv[indices[j],:] = np.copy(x[indices[j],:m])
    x_rc_tf = np.zeros((x.shape[0],m))
    for i in range(m):
        x_var = cp.Variable(G.N)

        obj = cp.Minimize(cp.sum_squares(x_mv[:,i] - x_var) + beta * cp.pnorm(D @ x_var, 1))
        prob = cp.Problem(obj)

        prob.solve(solver=cp.CVXOPT, verbose = False)

        if prob.status != cp.OPTIMAL:
            raise Exception("Solver did not converge!")

        x_rc_tf[:,i] = x_var.value

    rnmse_rc_tf[j] = np.sqrt(np.square(x_rc_tf[indices[j],:]-x[indices[j],:m]).mean())/x_mv[indices[j],:m].std()
    print(f"The rNMSE for {M[j]} observation is {np.round(rnmse_rc_tf[j],4)}.")
```

In the following cell, define a two layer MLP using the parameters given above:

```python
######################################################################### 
# define a two layer MLP
class MLP(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, dropout):
        super().__init__()
        self.in_channels = in_channels
        self.hidden_channels = hidden_channels
        self.out_channels = out_channels
        self.dropout = torch.nn.Dropout(p=dropout)
        self.layer1 = nn.Linear(in_channels, hidden_channels)
        self.layer2 = nn.Linear(hidden_channels, out_channels)
        self.activation = torch.nn.ReLU()

    def forward(self, x):
        x1 = self.activation(self.layer1(x))
        x1 = self.dropout(x1)
        out = self.layer2(x1)
        return out
#########################################################################   
```

In the following cell, train your MLP for 200 epochs and print the training loss, accuracy on the test set for each epoch:

```python
######################################################################### 
# instantiate your MLP model here
mlp1 = MLP(in_channels=3,
           hidden_channels=16,
           out_channels=7,
           dropout=0.0)
# instantiate an optimizer for the MLP
optimizer = torch.optim.Adam(list(mlp1.parameters()), lr=0.01)
######################################################################### 


######################################################################### 
# define a training function
def train():
    mlp1.train()
    optimizer.zero_grad()
    out = mlp1(features1)
    loss = torch.nn.functional.cross_entropy(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return float(loss)
######################################################################### 

######################################################################### 
# define a test function
@torch.no_grad()
def test():
    mlp1.eval()
    pred = mlp1(features1).argmax(dim=-1)

    accs = []
    for mask in [data.train_mask, data.val_mask, data.test_mask]:
        accs.append(int((pred[mask] == data.y[mask]).sum()) / int(mask.sum()))
    return accs
######################################################################### 

######################################################################### 
#train your model here and print the accuracy
train_acc_list, val_acc_list, test_acc_list = [], [], []
best_val_acc = final_test_acc = 0
for epoch in range(1, 200):
    loss = train()
    train_acc, val_acc, test_acc = test()
    train_acc_list.append(train_acc)
    val_acc_list.append(val_acc)
    test_acc_list.append(test_acc)
    print(f'Epoch: {epoch:03d}, Train: {train_acc:.4f}, Val: {val_acc:.4f}, Test: {test_acc:.4f}')
######################################################################### 
```

### ðŸ“š Decription of the dataset:    

Protein-protein interactions (PPI) are very important to discover new putative therapeutic targets to cure disease. Expensive and time-consuming wet-lab results are usually required to obtain PPI activity. PPI prediction aims to predict the PPI activity given a pair of proteins' amino acid sequences. For the Protein-protein interaction dataset, we build up the network as follows:

    - Each protein is a node on an undirected graph G. 
    - There is a connection between the two nodes if there is a known interaction in the training data.
    - The link prediction problem in this case is defined as: given the two nodes on the undirected graph, predict whether there is a link between them.

More information about the used PPI dataset can be found here: https://tdcommons.ai/multi_pred_tasks/ppi/

```python
######################################################################### 
# add your code here to generate embeddings for the PPI dataset using Node2Vec

model = Node2Vec(graph_pyg.edge_index, embedding_dim=3, walk_length=20,
                 context_size=10, walks_per_node=10, num_negative_samples=1,
                 p=1, q=1, sparse=True)
loader = model.loader(batch_size=128, shuffle=True,
                          num_workers=2)
optimizer = torch.optim.SparseAdam(list(model.parameters()), lr=0.01)

def train():
        model.train()
        total_loss = 0
        for pos_rw, neg_rw in loader:
            optimizer.zero_grad()
            loss = model.loss(pos_rw, neg_rw)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        return total_loss / len(loader)
for epoch in range(1, 101):
        loss = train()
        #acc = test()
        print(f'Epoch: {epoch:02d}, Loss: {loss:.4f}, '
              #Acc: {acc:.4f}'
             )
######################################################################### 
```
